## Apa itu Regresi?

**Regresi** adalah metode statistik dan machine learning untuk memodelkan hubungan antara satu atau lebih variabel input (**X/feature**) dengan variabel output (**Y/target**).

- **Tujuan utama:** Memprediksi nilai Y berdasarkan nilai X.
- **Contoh:** Prediksi harga rumah dari luas tanah, jumlah kamar, lokasi, dll.

### Jenis Regresi:
- **Regresi Linear:** Hubungan antara X dan Y berbentuk garis lurus.
- **Regresi Non-Linear:** Hubungan antara X dan Y tidak harus garis lurus (bisa kurva).
- **Regresi Berganda:** Menggunakan lebih dari satu fitur (X) untuk memprediksi Y.

### Aplikasi Regresi:
- Prediksi penjualan
- Prediksi suhu
- Prediksi permintaan sepeda
- Prediksi harga saham

**Kesimpulan:**  
Regresi digunakan untuk memprediksi nilai numerik (kontinu) dan memahami pengaruh fitur terhadap target.

## MSE, RMSE, dan R-Square

### 1. **Residual**
- **Residual = Y - Y-pred**
- Residual adalah selisih antara nilai aktual (**Y**) dan prediksi model (**Y-pred**).
- Contoh:
    | Y  | Y-Pred | Residual |
    |----|--------|----------|
    | 12 | 13     | -1       |
    | 14 | 12     | 2        |
    | 10 | 11     | -1       |
    | 11 | 12     | 1        |
    | 15 | 15     | 0        |
    | 16 | 17     | -1       |
    | 17 | 18     | -1       |
    | 14 | 12     | 2        |

---

### 2. **MSE (Mean Squared Error)**
- MSE adalah rata-rata dari kuadrat residual.
- Rumus:
  \[
  MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2
  \]
- Satuan hasil MSE adalah kuadrat dari satuan target.

---

### 3. **RMSE (Root Mean Squared Error)**
- RMSE adalah akar dari MSE.
- Rumus:
  \[
  RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2}
  \]
- Satuan hasil RMSE sama dengan satuan target (misal: sepeda, rupiah, dll).
- RMSE **sangat sensitif terhadap outlier** (error besar akan sangat mempengaruhi nilai RMSE).

---

### 4. **Kapan Menggunakan RMSE/MSE?**
- Variansi Y kecil (data tidak terlalu menyebar).
- Tidak ada outlier atau hanya sedikit outlier pada residual.
- Jika ingin penalti lebih besar untuk error yang besar.

---

### 5. **R-Square (R²)**
- R² mengukur seberapa baik model menjelaskan variabilitas data target.
- Nilai R² berkisar dari 0 sampai 1 (atau 0% sampai 100%).
- Semakin tinggi R², semakin baik model dalam menjelaskan data.

## Mean Absolute Percentage Error (MAPE)

- **MAPE** mengukur rata-rata error model dalam bentuk persentase terhadap nilai aktual.
- Rumus:
  \[
  MAPE = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{Y_i - \hat{Y}_i}{Y_i} \right| \times 100\%
  \]
- **Residual(%) = (Y - Y-pred) / Y**
- Contoh:
    | Y    | Y-Pred | Residual | Residual(%) |
    |------|--------|----------|-------------|
    | 1000 | 1050   | -50      | -5%         |
    | 100  | 150    | -50      | -50%        | ← nilai kecil, error persentase besar
    | 1000 | 900    | 100      | 10%         |
    | 110  | 120    | -10      | -9%         |
    | 150  | 150    | 0        | 0%          |
    | 770  | 870    | -100     | -13%        |
    | 760  | 780    | -20      | -3%         |
    | 650  | 350    | 300      | 46%         | ← outlier

- **Kapan menggunakan MAPE?**
  1. Jika ingin mengetahui error relatif (persentase) terhadap nilai aktual.
  2. Jika banyak outlier pada error relatif.

- **Catatan:**  
  - MAPE bisa sangat besar jika nilai Y kecil (lihat contoh -50%).
  - Outlier pada residual(%) sangat mempengaruhi MAPE.

---

## Mean Absolute Error (MAE)

- **MAE** mengukur rata-rata error absolut model dalam satuan asli target.
- Rumus:
  \[
  MAE = \frac{1}{n} \sum_{i=1}^{n} |Y_i - \hat{Y}_i|
  \]
- **Residual = Y - Y-pred**
- Contoh:
    | Y  | Y-Pred | Residual |
    |----|--------|----------|
    | 10 | 13     | -3       |
    | 14 | 10     | 4        |
    | 16 | 11     | 5        |
    | 12 | 11     | 1        |
    | 15 | 15     | 0        |
    | 16 | 17     | -1       |
    | 17 | 18     | -1       |
    | 14 | 32     | -16      | ← outlier

- **Kapan menggunakan MAE?**
  1. Jika variansi Y kecil.
  2. Jika banyak outlier pada residual.

- **Catatan:**  
  - MAE lebih robust terhadap outlier dibanding RMSE.
  - MAE mudah diinterpretasi karena satuannya sama dengan target.

## Root Mean Squared Percentage Error (RMSPE)

- **RMSPE** mengukur rata-rata error model dalam bentuk persentase, namun menggunakan kuadrat dari error persentase (mirip MAPE, tapi di-root dan di-square dulu).
- Rumus:
  \[
  RMSPE = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} \left( \frac{Y_i - \hat{Y}_i}{Y_i} \right)^2 }
  \]
- **Residual(%) = (Y - Y-pred) / Y**

### Contoh:
| Y    | Y-Pred | Residual | Residual(%) |
|------|--------|----------|-------------|
| 1000 | 1050   | -50      | -5%         |
| 250  | 300    | -50      | -20%        | ← residual sama, persentase beda
| 1000 | 900    | 100      | 10%         |
| 110  | 120    | -10      | -9%         |
| 770  | 870    | -100     | -13%        |
| 760  | 780    | -20      | -3%         |
| 650  | 560    | 90       | 14%         |
| 10   | 50     | -40      | -400%       | ← hati-hati jika Y sangat kecil

### Kapan menggunakan RMSPE?
1. Jika lebih tertarik pada error relatif (persentase) daripada error absolut.
2. Jika tidak ada atau sangat sedikit outlier pada residual absolut.

### Catatan:
- RMSPE **sangat sensitif terhadap outlier**, terutama jika nilai Y sangat kecil (lihat contoh -400%).
- Residual yang sama bisa menghasilkan persentase error yang sangat berbeda jika nilai Y berbeda.
- Gunakan RMSPE dengan hati-hati jika data memiliki nilai target yang kecil.

## Residual Analysis

Residual analysis digunakan untuk mengevaluasi apakah model regresi sudah sesuai dengan pola data.

### Penjelasan Gambar:

1. **Plot Kiri (Random/Good Residuals)**
   - Titik-titik residual tersebar acak di sekitar garis nol.
   - Tidak ada pola khusus.
   - **Artinya:** Model sudah sesuai, asumsi linearitas dan homoskedastisitas terpenuhi.

2. **Plot Tengah (Nonlinear Pattern)**
   - Residual membentuk pola melengkung (misal: parabola).
   - Ada pola sistematis, bukan acak.
   - **Artinya:** Model linear kurang tepat, mungkin data lebih cocok dengan model non-linear.

3. **Plot Kanan (Heteroskedastisitas)**
   - Residual semakin besar seiring naiknya nilai X (fan shape).
   - Variansi residual tidak konstan.
   - **Artinya:** Ada masalah heteroskedastisitas, model bisa bias dan prediksi kurang akurat pada nilai X tertentu.

---

**Kesimpulan:**  
- Residual yang baik = acak di sekitar nol (tidak ada pola).
- Jika ada pola, pertimbangkan model non-linear atau lakukan transformasi data.
- Jika variansi residual tidak konstan, pertimbangkan penanganan heteroskedastisitas.

````python
# library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, KFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# modeling
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score,
    root_mean_squared_error
)
from sklearn.pipeline import Pipeline

import warnings
warnings.filterwarnings('ignore')
````

````python
# data

tips = sns.load_dataset('tips')
tips.head()
````

````python
# data info
tips.info()
````

````python
# preprocessing
prepros = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), ['sex', 'smoker', 'day', 'time']),
], remainder='passthrough')
````

````python
# Data Splitting

x = tips.drop('tip', axis=1)
y = tips['tip']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, 
    test_size=0.2, 
    random_state=42
)
````

``` tidak perlu menambahkan stratified dibagian train test split, karena stratified hanya digunakan untuk klasifikasi guna menjaga proporsi kelas di train dan test tetap sama. ```

````python
# Define models

linreg = LinearRegression()
knn = KNeighborsRegressor(n_neighbors=5)
dt = DecisionTreeRegressor(random_state=42, max_depth=5)
rf = RandomForestRegressor(random_state=42, n_estimators=100, max_depth=5)
svm = SVR(kernel='rbf')

model = [linreg, knn, dt, rf, svm]
````

````python
# testing dengan model knn reg
knnPipe = Pipeline([
    ('preprocessing', prepros),
    ('model', knn)
])


score = cross_val_score(knnPipe, x_train, y_train, cv=5, scoring='neg_mean_absolute_percentage_error')
abs(score.mean())
````

````python
# Recap scroring (MAPE)

CVMAPE = []
avgMAPE = []
stdMAPE = []
CVRMSE = []
avgRMSE = []
stdRMSE = []
CVR2 = []
avgR2 = []
stdR2 = []


for i in model:
    pipe = Pipeline([
        ('preprocessing', prepros),
        ('model', i)
    ])
    scoreMAPE = cross_val_score(pipe, x_train, y_train, cv=5, scoring='neg_mean_absolute_percentage_error')
    scoreRMSE = cross_val_score(pipe, x_train, y_train, cv=5, scoring='neg_root_mean_squared_error')
    scoreR2 = cross_val_score(pipe, x_train, y_train, cv=5, scoring='r2')
    CVMAPE.append(scoreMAPE)
    avgMAPE.append(abs(scoreMAPE.mean()))
    stdMAPE.append(scoreMAPE.std())
    CVRMSE.append(scoreRMSE)
    avgRMSE.append(abs(scoreRMSE.mean()))
    stdRMSE.append(scoreRMSE.std())
    CVR2.append(scoreR2)
    avgR2.append(scoreR2.mean())
    stdR2.append(scoreR2.std())

summary = pd.DataFrame({
    'Model': ['LinearRegression', 'KNeighborsRegressor', 'DecisionTreeRegressor', 'RandomForestRegressor', 'SVR'],
    'Avg MAPE': np.round(np.abs(avgMAPE), 4),
    'Std MAPE': np.round(stdMAPE, 4),
    'Avg RMSE': np.round(np.abs(avgRMSE), 4),
    'Std RMSE': np.round(stdRMSE, 4),
    'Avg R2 Score': np.round(avgR2, 4),
    'Std R2 Score': np.round(stdR2, 4)
}).sort_values(by='Avg MAPE', ascending=False)

summary
````

# Study Case: Predict Customer Lifetime Value (LTV)

````python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
````

````python
# 1. Load data
clv = pd.read_csv('CLV.csv')

clv.head()
````

````python
clv.info()
````

````python
# 2.Cleaning & preprocessing  
# Normalisasi teks
clv['subscription_plan'] = clv['subscription_plan'].str.strip().str.lower().map({'basic':'Basic','standard':'Standard','premium':'Premium'})
clv['device_type'] = clv['device_type'].str.strip().str.lower().map({'mobile':'Mobile','desktop':'Desktop'})

# Handling missing
if 'avg_watch_time' in clv.columns:
    clv['avg_watch_time'] = clv['avg_watch_time'].fillna(clv['avg_watch_time'].median())

if 'region' in clv.columns:
    clv['region'] = clv['region'].fillna('Unknown')

# 3. Features & target
target = 'customer_lifetime_value'
features = [c for c in clv.columns if c != target and c != 'customer_id']

x = clv[features]
y = clv[target]
````

````python
# 4. Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 5. Preprocessing
cat_cols = [c for c in x_train.columns if x_train[c].dtype == 'object']
prepros_clv = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols)
], remainder='passthrough')
````

````python
# 6. Models
linreg = LinearRegression()
knn = KNeighborsRegressor(n_neighbors=5)
dt = DecisionTreeRegressor(random_state=42, max_depth=6)
rf = RandomForestRegressor(random_state=42, n_estimators=150, max_depth=6)
svm = SVR(kernel='rbf', C=1.0, epsilon=0.1)
models = [linreg, knn, dt, rf, svm]
model_names = ['LinearRegression', 'KNeighborsRegressor', 'DecisionTreeRegressor', 'RandomForestRegressor', 'SVR']
````

````python
# 7. Metric 
CVMAE, avgMAE, stdMAE = [], [], []
CVMAPE, avgMAPE, stdMAPE = [], [], []
CVR2, avgR2, stdR2 = [], [], []

for m in models:
    pipe = Pipeline([
        ('preprocessing', prepros_clv),
        ('model', m)
    ])
    # MAE
    scoreMAE = cross_val_score(pipe, x_train, y_train, cv=5, scoring='neg_mean_absolute_error')
    # MAPE
    scoreMAPE = cross_val_score(pipe, x_train, y_train, cv=5, scoring='neg_mean_absolute_percentage_error')
    # R2
    scoreR2 = cross_val_score(pipe, x_train, y_train, cv=5, scoring='r2')

    CVMAE.append(scoreMAE)
    avgMAE.append(abs(scoreMAE.mean()))
    stdMAE.append(scoreMAE.std())

    CVMAPE.append(scoreMAPE)
    avgMAPE.append(abs(scoreMAPE.mean()))
    stdMAPE.append(scoreMAPE.std())

    CVR2.append(scoreR2)
    avgR2.append(scoreR2.mean())
    stdR2.append(scoreR2.std())

# 8. Summary
summary_clv = pd.DataFrame({
    'Model': model_names,
    'Avg MAE': np.round(avgMAE, 4),
    'Std MAE': np.round(stdMAE, 4),
    'Avg MAPE': np.round(avgMAPE, 4),
    'Std MAPE': np.round(stdMAPE, 4),
    'Avg R2': np.round(avgR2, 4),
    'Std R2': np.round(stdR2, 4)
}).sort_values(by='Avg MAE', ascending=True)

summary_clv
````

````python
best_model_name = summary_clv.iloc[0]['Model']
best_mae = summary_clv.iloc[0]['Avg MAE']
best_mape = summary_clv.iloc[0]['Avg MAPE']
best_r2 = summary_clv.iloc[0]['Avg R2']

# Visualisasi
plt.figure(figsize=(10, 6))
plt.barh(summary_clv['Model'], summary_clv['Avg MAE'])
plt.xlabel('Average MAE (Rupiah)')
plt.title('Model Comparison: Average MAE')
plt.tight_layout()
plt.show()

# Interpretasi bisnis
print(f"Model terbaik: {best_model_name}")
print(f"Rata-rata error (MAE): Rp {best_mae:,.0f}")
print(f"Rata-rata error (MAPE): {best_mape:.2%}")
print(f"R² Score: {best_r2:.4f}")
print(f"\nArtinya: Prediksi LTV meleset rata-rata sebesar Rp {best_mae:,.0f}")
````
