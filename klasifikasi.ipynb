# Klasifikasi
Klasifikasi adalah salah satu teknik dalam machine learning yang digunakan untuk mengelompokkan data ke dalam kategori atau kelas tertentu berdasarkan fitur-fitur yang dimilikinya. Tujuan utama dari klasifikasi adalah untuk memprediksi label atau kategori dari data baru berdasarkan pola yang telah dipelajari dari data pelatihan.
Respon variable = fungsi model + error acak
- Y = f(X) + e
- Y Kategorikal - 2 kategori (biner klasifikasi)
- Y Kategorikal - lebih dari 2 kategori (multikelas klasifikasi)

## Regresi logistik
Regresi logistik adalah salah satu algoritma klasifikasi yang digunakan untuk memprediksi probabilitas suatu kejadian yang bersifat biner (dua kelas). Meskipun namanya mengandung kata "regresi", regresi logistik sebenarnya adalah metode klasifikasi, bukan regresi dalam arti tradisional.
Regresi logistik digunakan ketika variabel target (respon) adalah kategorikal dengan dua kelas, misalnya:
- Ya/Tidak
- Sukses/Gagal
- 1/0

## Pohon keputusan
Pohon keputusan (decision tree) adalah salah satu algoritma machine learning yang digunakan untuk tugas klasifikasi dan regresi. Algoritma ini membangun model prediktif dalam bentuk struktur pohon, di mana setiap node internal mewakili sebuah fitur (atribut), setiap cabang mewakili keputusan berdasarkan nilai fitur tersebut, dan setiap daun (leaf) mewakili hasil prediksi atau kelas akhir.
Pohon keputusan bekerja dengan cara membagi data menjadi subset-subset yang lebih kecil berdasarkan fitur-fitur yang paling informatif. Proses ini berlanjut hingga mencapai kondisi berhenti tertentu, seperti kedalaman maksimum pohon atau jumlah minimum sampel di setiap daun.

# K-nearest Neighbors (KNN)
K-nearest Neighbors (KNN) adalah algoritma machine learning yang digunakan untuk tugas klasifikasi dan regresi. KNN adalah algoritma berbasis instance, yang berarti bahwa ia tidak membangun model eksplisit selama fase pelatihan, melainkan menyimpan seluruh dataset pelatihan dan menggunakan data tersebut untuk membuat prediksi pada data baru.
Prinsip dasar KNN adalah sebagai berikut:
1. Menentukan nilai K, yaitu jumlah tetangga terdekat yang akan dipertimbangkan untuk membuat prediksi.
2. Menghitung jarak antara data baru dan semua data dalam dataset pelatihan menggunakan metrik jarak tertentu (misalnya, Euclidean, Manhattan, atau Minkowski).
3. Memilih K tetangga terdekat berdasarkan jarak yang telah dihitung.
4. Untuk klasifikasi, menentukan kelas mayoritas dari K tetangga terdekat dan menetapkan kelas tersebut sebagai prediksi untuk data baru. Untuk regresi, menghitung rata-rata nilai target dari K tetangga terdekat dan menetapkan nilai tersebut sebagai prediksi.

# Super Vector Machine (SVM)
Support Vector Machine (SVM) adalah algoritma machine learning yang digunakan untuk tugas klasifikasi dan regresi. SVM dirancang untuk menemukan hyperplane optimal yang memisahkan data dari dua kelas dengan margin terbesar. Hyperplane adalah garis (dalam 2D), bidang (dalam 3D), atau ruang berdimensi lebih tinggi yang memisahkan kelas-kelas dalam dataset.
Prinsip dasar SVM adalah sebagai berikut:
1. Mencari hyperplane yang memaksimalkan margin antara dua kelas. Margin adalah jarak antara hyperplane dan titik data terdekat dari masing-masing kelas, yang disebut support vectors.
2. Jika data tidak dapat dipisahkan secara linear, SVM menggunakan teknik yang disebut kernel trick untuk memetakan data ke ruang berdimensi lebih tinggi di mana pemisahan linear mungkin terjadi 
3. Setelah hyperplane optimal ditemukan, SVM dapat digunakan untuk memprediksi kelas dari data baru berdasarkan posisi relatifnya terhadap hyperplane.

# Binary Logistic Regression
adalah model klasifikasi yang digunakan untuk memprediksi probabilitas kejadian dari dua kelas atau kategori yang saling eksklusif. Model ini digunakan ketika variabel target (respon) bersifat biner, seperti "ya/tidak", "sukses/gagal", atau "1/0". Logistic regression mengestimasi hubungan antara satu atau lebih variabel independen (fitur) dan variabel dependen biner dengan menggunakan fungsi logit.
Secara konseptual, logistic regression memodelkan probabilitas kejadian suatu peristiwa (misalnya, kelas "1") sebagai fungsi dari variabel independen. Fungsi logit mengubah probabilitas ini menjadi nilai yang dapat berkisar dari negatif tak terhingga hingga positif tak terhingga, yang kemudian dapat diubah kembali menjadi probabilitas menggunakan fungsi sigmoid.
Probabilitas dihitung menggunakan rumus berikut:
P(Y=1|X) = 1 / (1 + e^-(β0 + β1X1 + β2X2 + ... + βnXn))
Di mana:
- P(Y=1|X) adalah probabilitas kejadian kelas "1" diberikan fitur X
- β0 adalah intercept (konstanta)
- β1, β2, ..., βn adalah koefisien regresi untuk masing-masing fitur X1, X2, ..., Xn
- e adalah basis dari logaritma natural (sekitar 2.71828)

- 1 = Positif class
- 0 = Negatif class
- [ secara umum ]

## menentukan kelas positif dan negatif
- fraud / not fraud
- cancel / not cancel
- churn / not churn
- terima pinjama / tolak pinjaman
- dll


Output sebenernya adalah probabilitas, untuk mengklasifikasikan ke dalam kelas 0 atau 1, kita perlu menentukan threshold (ambang batas). Secara umum, threshold yang umum digunakan adalah 0.5. Artinya:
- Jika probabilitas >= 0.5, maka prediksi adalah kelas 1 (positif)
- Jika probabilitas < 0.5, maka prediksi adalah kelas 0 (negatif)


## Kenapa kita gunakan regresi logistik daripada regresi linear?
- Regresi linear menghasilkan output kontinu yang dapat berada di luar rentang 0 hingga 1, sedangkan regresi logistik menghasilkan output dalam bentuk probabilitas yang selalu berada di antara 0 dan 1.
- Regresi logistik menggunakan fungsi logit (sigmoid) untuk memetakan output ke rentang probabilitas, sehingga lebih sesuai untuk masalah klasifikasi biner.
- Regresi logistik mengoptimalkan fungsi likelihood, yang lebih sesuai untuk klasifikasi, sedangkan regresi linear mengoptimalkan fungsi kesalahan kuadrat, yang lebih cocok untuk regresi kontinu.
- Regresi logistik lebih tahan terhadap outlier dalam konteks klasifikasi, karena fokusnya pada probabilitas kelas, sedangkan regresi linear dapat terpengaruh secara signifikan oleh outlier dalam data.

- Odds(Y = 1 | Female) = 0.10 / 0.90 = 1/9 ≈ 0.11

### Odds Ratio (Male vs Female)

```python
# Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split # Gunakan untuk membuat 2 jenis data: training dan testing bertujuan untuk mengetahui apakah model yang dibuat sudah baik atau belum
from sklearn.linear_model import LogisticRegression # Gunakan untuk membuat model klasifikasi dengan algoritma regresi logistik
from sklearn.metrics import accuracy_score # Gunakan untuk mengukur akurasi dari model klasifikasi

import warnings
warnings.filterwarnings('ignore')
```

# K-nearest neighbors (KNN)
K-nearest Neighbors (KNN) adalah algoritma machine learning yang digunakan untuk tugas klasifikasi dan regresi. KNN adalah algoritma berbasis instance, yang berarti bahwa ia tidak membangun model eksplisit selama fase pelatihan, melainkan menyimpan seluruh dataset pelatihan dan menggunakan data tersebut untuk membuat prediksi pada data baru.
Prinsip dasar KNN adalah sebagai berikut:
1. Menentukan nilai K, yaitu jumlah tetangga terdekat yang akan dipertimbangkan untuk membuat prediksi.
2. Menghitung jarak antara data baru dan semua data dalam dataset pelatihan menggunakan metrik jarak tertentu (misalnya, Euclidean, Manhattan, atau Minkowski).
3. Memilih K tetangga terdekat berdasarkan jarak yang telah dihitung.
4. Untuk klasifikasi, menentukan kelas mayoritas dari K tetangga terdekat dan menetapkan kelas tersebut sebagai prediksi untuk data baru. Untuk regresi, menghitung rata-rata nilai target dari K tetangga terdekat dan menetapkan nilai tersebut sebagai prediksi.

## cara memilih nilai K
- Nilai K yang kecil (misalnya, K=1) dapat membuat model sangat sensitif terhadap noise dalam data, sehingga dapat menyebabkan overfitting.
- Nilai K yang besar (misalnya, K=20) dapat membuat model terlalu umum dan mengabaikan pola lokal dalam data, sehingga dapat menyebabkan underfitting.
- Nilai K yang ganjil sering dipilih untuk menghindari situasi di mana terjadi seri dalam pemungutan suara untuk klasifikasi biner.

### Ecluidean Distance
Jarak Euclidean adalah metrik yang paling umum digunakan dalam KNN untuk mengukur jarak antara dua titik dalam ruang fitur. Rumus jarak Euclidean antara dua titik A (x1, y1) dan B (x2, y2) dalam ruang dua dimensi adalah sebagai berikut:
d(A, B) = √((x2 - x1)² + (y2 - y1)²)
Untuk ruang berdimensi lebih tinggi, rumusnya dapat diperluas menjadi:
d(A, B) = √(Σ (xi - yi)²)
Di mana:
- d(A, B) adalah jarak Euclidean antara titik A dan B
- xi dan yi adalah nilai fitur dari titik A dan B pada dimensi ke-i
- Σ menunjukkan penjumlahan atas semua dimensi fitur

### Measuring Distance
| Distance Metric     | Formula                                                                 |
|---------------------|-------------------------------------------------------------------------|
| Euclidean Distance  | d(A, B) = √(Σ (xi - yi)²)                                      |
| Manhattan Distance  | d(A, B) = Σ |xi - yi|                                      |
| Minkowski Distance  | d(A, B) = (Σ |xi - yi|^p)^(1/p)                             |
| Hamming Distance    | d(A, B) = Σ (xi ≠ yi) (for categorical variables)                  |    
| Cosine Similarity   | sim(A, B) = (A · B) / (||A|| ||B||) (not a distance but a similarity measure) |
